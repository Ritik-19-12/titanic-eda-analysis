```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/Titanic.csv')  
df.head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='pastel')
plt.title("Survival by Sex")
plt.savefig('../visuals/relationship_survival_sex.png')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='muted')
plt.title("Survival by Passenger Class")
plt.savefig('../visuals/relationship_survival_pclass.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True, palette='Set2')
plt.title("Age Distribution by Survival")
plt.savefig('../visuals/relationship_age_survival.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='Survived', y='Fare', data=df, palette='coolwarm')
plt.title("Fare Distribution by Survival")
plt.savefig('../visuals/relationship_fare_survival.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.savefig('../visuals/correlation_heatmap.png')
plt.show()
```
